Loading Clean Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\A.K\DATA_SCIENCE_WORKSPACE\kaggle_projects\project_2_sales_analysis\data\processed\cleaned_sales_data.xlsx", index_col=0)

df.head()
```

CHECKING SALES TREND OVER TIME

```python
# Monthly sales trend
monthly_sales = df.resample('ME', on='order_date')['sales'].sum()

#PLOTTED IN A LINE PLOT USING SEABORN
monthly_sales.plot(figsize=(12,6), title="Monthly Sales Trend")
plt.ylabel("Sales")

plt.savefig("C:/Users/A.K/DATA_SCIENCE_WORKSPACE/kaggle_projects/project_2_sales_analysis/outputs/figures/monthly_sales_trend.png")
plt.show()
```

SALES BY REGION

```python
#PLOTTED IN A BAR PLOT USING SEABORN
region_sales = df.groupby('region')['sales'].sum().sort_values()

region_sales.plot(kind='barh', title="Sales by Region")

plt.savefig("C:/Users/A.K/DATA_SCIENCE_WORKSPACE/kaggle_projects/project_2_sales_analysis/outputs/figures/sales_by_region.png")
plt.show()
```

```python
#profit by region
region_profit = df.groupby('region')['profit'].sum().sort_values()

region_profit.plot(kind='barh', title="Profit by Region")

plt.savefig("C:/Users/A.K/DATA_SCIENCE_WORKSPACE/kaggle_projects/project_2_sales_analysis/outputs/figures/profit_by_region.png")
plt.show()
```

PROFIT BY CATEGORY

```python
#PLOTTED IN A BAR PLOT USING SEABORN
category_profit = df.groupby('category')['profit'].sum()

category_profit.plot(kind='bar', title="Profit by Category")

plt.savefig("C:/Users/A.K/DATA_SCIENCE_WORKSPACE/kaggle_projects/project_2_sales_analysis/outputs/figures/profit_by_category.png")
plt.show()
```

TOP LOSS-MAKING PRODUCTS

```python
#PLOTTED IN A BAR PLOT USING SEABORN
top_losses = df.groupby('product_name')['profit'].sum().sort_values().head(10)

top_losses.plot(kind='barh', title="Top 10 Loss-Making Products")

plt.savefig("C:/Users/A.K/DATA_SCIENCE_WORKSPACE/kaggle_projects/project_2_sales_analysis/outputs/figures/top_losses.png")  
plt.show()
```

TOP CUSTOMERS

```python
#PLOTTED IN A BAR PLOT USING SEABORN
top_customers = df.groupby('customer_name')['sales'].sum().sort_values(ascending=False).head(10)

top_customers.plot(kind='barh', title="Top Customers by Sales")

plt.savefig("C:/Users/A.K/DATA_SCIENCE_WORKSPACE/kaggle_projects/project_2_sales_analysis/outputs/figures/top_customers.png")   
plt.show()
```

SALES VS PROFIT (RELATIONSHIP)

```python
# Scatter plot of sales vs profit
plt.figure(figsize=(8,6))
sns.scatterplot(x='sales', y='profit', data=df)
plt.title("Sales vs Profit")

plt.savefig("C:/Users/A.K/DATA_SCIENCE_WORKSPACE/kaggle_projects/project_2_sales_analysis/outputs/figures/sales_vs_profit.png") 
plt.show()
```
