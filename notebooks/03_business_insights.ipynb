# Global Superstore – Business Insights Report

## Objective
This notebook translates EDA findings into business insights and actionable recommendations for decision-making.

## In Summary

- The business shows strong long-term sales growth with seasonal dips in January
- Profit is highly concentrated in the Technology category
- Central region is the strongest profit contributor
- Several regions and products are consistently underperforming or loss-making
- A small number of products are responsible for significant losses

## Business Context

The goal is to identify 'Key Drivers of Profit and Areas of Business Incompetence.
This analysis is based on the Global Superstore dataset and focuses on:

1. Revenue performance over time
2. Profitability by region and category
3. Product-level losses
4. Customer and market performance



## Insights By Business Area

### i) Sales Trend Analysis

- Sales show a strong upward trend over time
- Higher highs and higher lows indicate consistent growth
- January consistently shows seasonal dips
- Post-2014 period shows accelerated growth and new peak levels

### ii) Regional Profit Performance

- Central region is the highest profit contributor
- Significant performance gap exists between top and bottom regions
- Canada and Caribbean regions are underperforming
- Profit distribution is highly uneven across markets
